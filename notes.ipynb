let us create a pandas df and then use matplotlib to plot a graph

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "Quater" : ["Q1 2022", "Q2 2022", "Q3 2022", "Q4 2022", "Q1 2023" ],
    "Fridge" : [12, 11, 13, 16, 13],
    "Dishwasher" : [ 8, 7, 9, 11, 9],
    "Washing Machine" : [9, 8, 7, 7, 6]
})

df
```

Use the above data to plot a graphs:

- Line Chart

```python
from matplotlib import pyplot as plt

plt.plot(df["Quater"], df["Fridge"])
plt.show()
```

Increase the size of the chart

```python
plt.figure(figsize=(12,4)) #(width, height)
plt.plot(df["Quater"], df["Fridge"])
plt.show()
```

Lets us do the same for the other products

```python

plt.figure(figsize=(12,4)) #(width, height)

# Plotting multiple lines on the same graphb
plt.plot(df["Quater"], df["Fridge"], color="blue", label="Fridge") 
plt.plot(df["Quater"], df["Dishwasher"], color="green", label="Dishwasher")
plt.plot(df["Quater"], df["Washing Machine"], color="purple", label="Washing Machine")

# Add title at the top
plt.title("Product Sales")

# Add labels on x and y axis
plt.xlabel("Financial Quater")
plt.ylabel("Revenue in Million ($)")

plt.legend() # Show the legend

plt.show() # Display the plot
```

- Pie Chart

Let us calculate the sum of each product (cols) and plot a pie chart

```python
total_sales = df[["Fridge", "Dishwasher", "Washing Machine"]].sum()
total_sales
```

```python
plt.pie(total_sales, labels=total_sales.index, autopct="%1.1f%%") # autopct to show percentage on the pie chart
plt.title("Total Sales Distribution")

plt.show()
```

Lets look at some ways to style our pie chart, we have many parameters :
- `explode=()`-> to move part ourward 

```python
plt.pie(total_sales, labels=total_sales.index, autopct="%1.1f%%", explode=(0.1, 0, 0)) # explode to pull out the first slice of the pie chart
plt.title("Total Sales Distribution")

plt.show()
```

- `shadow=True`-> to display shadow

```python
plt.pie(total_sales, labels=total_sales.index, autopct="%1.1f%%", explode=(0.1, 0, 0), shadow=True) # shadow to add shadow effect to the pie chart
plt.title("Total Sales Distribution")

plt.show()
```

- `startangle=` -> to adjust the angle 

```python
plt.pie(total_sales, labels=total_sales.index, autopct="%1.1f%%", startangle=90) # startangle to rotate the pie chart 
plt.title("Total Sales Distribution")

plt.show()
```

- Bar Chart

```python
df.plot(kind="bar", x="Quater", )

# Tilt the x-axis labels for better readability
plt.xticks(rotation=45)
plt.show()
```
